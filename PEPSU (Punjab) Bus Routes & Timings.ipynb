```python
#PEPSU (Punjab) Bus Routes & Timings


from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

driver = webdriver.Chrome()


time.sleep(5)

driver.get('https://www.redbus.in/online-booking/pepsu/?utm_source=rtchometile')


wait = WebDriverWait(driver, 10)


body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
body.send_keys(Keys.PAGE_DOWN)


PEPSU_bus_routes = []

for page_number in range(1, 4):
    try:
      
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[class='route']")))

   
        transport_name_elements = driver.find_elements(By.CSS_SELECTOR, "a[class='route']")

      
        for element in transport_name_elements:
            route = element.text.strip()
            if route and route != 'bus_routes':
                PEPSU_bus_routes.append(route)

        if page_number < 3:
       
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            
            
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            
            driver.maximize_window()
            actions = ActionChains(driver)                                       
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
         
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route in PEPSU_bus_routes:
    print(route)

driver.quit()
```

```python
#PEPSU BUS ROUTE LINK
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


driver = webdriver.Chrome()

time.sleep(5)


driver.get('https://www.redbus.in/online-booking/pepsu/?utm_source=rtchometile')

wait = WebDriverWait(driver, 10)
body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))


driver.maximize_window()

body.send_keys(Keys.PAGE_DOWN)
time.sleep(2)


PEPSU_bus_route_links = []


for page_number in range(1, 4):
    try:
       
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='route_link']")))

        transport_name_elements = driver.find_elements(By.XPATH, "//div[@class='route_link']")
        
      
        for element in transport_name_elements:
            link_element = element.find_element(By.TAG_NAME, 'a')
            route_link = link_element.get_attribute('href')  
            PEPSU_bus_route_links.append(route_link)

       
        if page_number < 3:
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route_link in PEPSU_bus_route_links:
    print(f"Link: {route_link}")


driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.implicitly_wait(10)


urls = [
    {'route': 'Patiala to Delhi', 'url': 'https://www.redbus.in/bus-tickets/patiala-to-delhi'},
    {'route': 'Delhi to Patiala', 'url': 'https://www.redbus.in/bus-tickets/delhi-to-patiala'},
    {'route': 'Ludhiana to Delhi', 'url': 'https://www.redbus.in/bus-tickets/ludhiana-to-delhi'},
    {'route': 'Delhi to Ludhiana', 'url': 'https://www.redbus.in/bus-tickets/delhi-to-ludhiana'},
    {'route': 'Phagwara to Delhi', 'url': 'https://www.redbus.in/bus-tickets/phagwara-to-delhi'},
    {'route': 'Jalandhar to Delhi', 'url': 'https://www.redbus.in/bus-tickets/jalandhar-to-delhi'},
    {'route': 'Delhi to Jalandhar', 'url': 'https://www.redbus.in/bus-tickets/delhi-to-jalandhar'},
    {'route': 'Patiala to Delhi Airport', 'url': 'https://www.redbus.in/bus-tickets/patiala-to-delhi-airport'},
    {'route': 'Jalandhar to Delhi Airport', 'url': 'https://www.redbus.in/bus-tickets/jalandhar-to-delhi-airport'},
    {'route': 'Ludhiana to Delhi Airport', 'url': 'https://www.redbus.in/bus-tickets/ludhiana-to-delhi-airport'},
    {'route': 'Phagwara to Delhi Airport', 'url': 'https://www.redbus.in/bus-tickets/phagwara-to-delhi-airport'},
    {'route': 'Delhi Airport to Ludhiana', 'url': 'https://www.redbus.in/bus-tickets/delhi-airport-to-ludhiana'},
    {'route': 'Delhi to Phagwara', 'url': 'https://www.redbus.in/bus-tickets/delhi-to-phagwara'},
    {'route': 'Delhi to Amritsar', 'url': 'https://www.redbus.in/bus-tickets/delhi-to-amritsar'},
    {'route': 'Amritsar to Delhi', 'url': 'https://www.redbus.in/bus-tickets/amritsar-to-delhi'},
    {'route': 'Delhi Airport to Patiala', 'url': 'https://www.redbus.in/bus-tickets/delhi-airport-to-patiala'},
    {'route': 'Amritsar to Delhi Airport', 'url': 'https://www.redbus.in/bus-tickets/amritsar-to-delhi-airport'},
    {'route': 'Kapurthala to Delhi', 'url': 'https://www.redbus.in/bus-tickets/kapurthala-to-delhi'},
    {'route': 'Delhi Airport to Jalandhar', 'url': 'https://www.redbus.in/bus-tickets/delhi-airport-to-jalandhar'},
    {'route': 'Chandigarh to Bathinda', 'url': 'https://www.redbus.in/bus-tickets/chandigarh-to-bathinda'},
    {'route': 'Chandigarh to Faridkot', 'url': 'https://www.redbus.in/bus-tickets/chandigarh-to-faridkot'},
    {'route': 'Chandigarh to Patiala', 'url': 'https://www.redbus.in/bus-tickets/chandigarh-to-patiala'}
]

all_buses = []

for route in urls:
    route_name = route['route']
    route_url = route['url']

    # Open the target URL
    driver.get(route_url)
    driver.maximize_window()
    time.sleep(5)

    # Scroll through the page until the end
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source

        for i in range(10):  
            body = driver.find_element(By.TAG_NAME, "body")
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1) 

        time.sleep(2)

        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

    # Use explicit wait to ensure the bus details elements are present in the DOM
    try:
        bus_details = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, ".//div[@class='clearfix bus-item']"))
        )
    except Exception as e:
        print(f"Error: {e} for route {route_name}")
        continue

    for bus in bus_details:
        try:
            bus_name = bus.find_element(By.XPATH, ".//div[@class='travels lh-24 f-bold d-color']").text
        except:
            bus_name = "N/A"
        try:
            departure_time = bus.find_element(By.XPATH, ".//div[@class='dp-time f-19 d-color f-bold']").text
        except:
            departure_time = "N/A"
        try:
            reaching_time = bus.find_element(By.XPATH, ".//div[@class='bp-time f-19 d-color disp-Inline']").text
        except:
            reaching_time = "N/A"
        try:
            duration = bus.find_element(By.XPATH, ".//div[@class='dur l-color lh-24']").text
        except:
            duration = "N/A"
        try:
            bus_price = bus.find_element(By.CSS_SELECTOR, "div[class='seat-fare ']").text
        except:
            bus_price = "N/A"
        try:
            seat_availability = bus.find_element(By.CSS_SELECTOR, "div[class='column-eight w-15 fl']").text
        except:
            seat_availability = "N/A"
        try:
            bus_type = bus.find_element(By.XPATH, ".//div[@class='bus-type f-12 m-top-16 l-color evBus']").text
        except:
            bus_type = "N/A"
        try:
            bus_rating = bus.find_element(By.XPATH, ".//div[@class='rating-sec lh-24']").text
        except:
            bus_rating = "N/A"

        all_buses.append({
            "Bus Name": bus_name,
            "Departure Time": departure_time,
            "Reaching Time": reaching_time,
            "Duration": duration,
            "Bus Price": bus_price,
            "Seat Availability": seat_availability,
            "Bus Type": bus_type,
            "Rating": bus_rating,
            "Bus Route": route_name,
            "Bus Route Link": route_url
        })

df = pd.DataFrame(all_buses)
print(df)

driver.quit()
```
