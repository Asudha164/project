```python
#Bihar state road transport corporation bus routes

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

driver = webdriver.Chrome()


time.sleep(5)

driver.get('https://www.redbus.in/online-booking/bihar-state-road-transport-corporation-bsrtc/?utm_source=rtchometile')


wait = WebDriverWait(driver, 10)


body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
body.send_keys(Keys.PAGE_DOWN)


bus_routes = []

for page_number in range(1, 5):
    try:
      
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[class='route']")))

   
        transport_name_elements = driver.find_elements(By.CSS_SELECTOR, "a[class='route']")

      
        for element in transport_name_elements:
            route = element.text.strip()
            if route and route != 'bus_routes':
                bus_routes.append(route)

        if page_number < 4:
       
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            
            
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            
            driver.maximize_window()
            actions = ActionChains(driver)                                       
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
         
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route in bus_routes:
    print(route)

driver.quit()
```

```python
#Bihar state road transport corporation bus routes LINK
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


driver = webdriver.Chrome()

time.sleep(5)


driver.get('https://www.redbus.in/online-booking/bihar-state-road-transport-corporation-bsrtc/?utm_source=rtchometile')

wait = WebDriverWait(driver, 10)
body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))


driver.maximize_window()

body.send_keys(Keys.PAGE_DOWN)
time.sleep(2)


bus_route_links = []


for page_number in range(1, 5):
    try:
       
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='route_link']")))

        transport_name_elements = driver.find_elements(By.XPATH, "//div[@class='route_link']")
        
      
        for element in transport_name_elements:
            link_element = element.find_element(By.TAG_NAME, 'a')
            route_link = link_element.get_attribute('href')  
            bus_route_links.append(route_link)

       
        if page_number < 4:
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route_link in bus_route_links:
    print(f"Link: {route_link}")


driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.implicitly_wait(10)

urls = [
    {'route': 'Patna to Bettiah', 'url': 'https://www.redbus.in/bus-tickets/patna-to-bettiah'},
    {'route': 'Gopalganj to Delhi', 'url': 'https://www.redbus.in/bus-tickets/gopalganj-to-delhi'},
    {'route': 'Patna to Motihari', 'url': 'https://www.redbus.in/bus-tickets/patna-to-motihari'}, 
    {'route': 'Delhi to Motihari', 'url': 'https://www.redbus.in/bus-tickets/delhi-to-motihari'},
    {'route': 'Motihari to Delhi', 'url': 'https://www.redbus.in/bus-tickets/motihari-to-delhi'},
    {'route': 'Patna to Balmiki Nagar Bihar', 'url': 'https://www.redbus.in/bus-tickets/patna-to-balmiki-nagar-bihar'},
    {'route': 'Balmiki Nagar Bihar to Patna', 'url': 'https://www.redbus.in/bus-tickets/balmiki-nagar-bihar-to-patna'},
    {'route': 'Patna to Kathmandu', 'url': 'https://www.redbus.in/bus-tickets/patna-to-kathmandu'},
    {'route': 'Patna to Katihar', 'url': 'https://www.redbus.in/bus-tickets/patna-to-katihar'},
    {'route': 'Patna to Purnea', 'url': 'https://www.redbus.in/bus-tickets/patna-to-purnea'},
    {'route': 'Patna to Hazaribagh', 'url': 'https://www.redbus.in/bus-tickets/patna-to-hazaribagh'}, 
    {'route': 'Ranchi to Patna', 'url': 'https://www.redbus.in/bus-tickets/ranchi-to-patna'}, 
    {'route': 'Hazaribagh to Patna', 'url': 'https://www.redbus.in/bus-tickets/hazaribagh-to-patna'}, 
    {'route': 'Patna to Raxaul', 'url': 'https://www.redbus.in/bus-tickets/patna-to-raxaul'},
    {'route': 'Muzaffarpur to Kathmandu', 'url': 'https://www.redbus.in/bus-tickets/muzaffarpur-to-kathmandu'}, 
    {'route': 'Patna to Ranchi', 'url': 'https://www.redbus.in/bus-tickets/patna-to-ranchi'},
    {'route': 'Muzaffarpur to Ranchi', 'url': 'https://www.redbus.in/bus-tickets/muzaffarpur-to-ranchi'},
    {'route': 'Kathmandu to Patna', 'url': 'https://www.redbus.in/bus-tickets/kathmandu-to-patna'},
    {'route': 'Ranchi to Muzaffarpur', 'url': 'https://www.redbus.in/bus-tickets/ranchi-to-muzaffarpur'},
    {'route': 'Motihari to Lucknow', 'url': 'https://www.redbus.in/bus-tickets/motihari-to-lucknow'},
    {'route': 'Lucknow to Motihari', 'url': 'https://www.redbus.in/bus-tickets/lucknow-to-motihari'},
    {'route': 'Motihari to Kathmandu', 'url': 'https://www.redbus.in/bus-tickets/motihari-to-kathmandu'},
    {'route': 'Agra to Motihari', 'url': 'https://www.redbus.in/bus-tickets/agra-to-motihari'},
    {'route': 'Patna to Janakpur Nepal', 'url': 'https://www.redbus.in/bus-tickets/patna-to-janakpur-nepal'},
    {'route': 'Muzaffarpur to Hazaribagh', 'url': 'https://www.redbus.in/bus-tickets/muzaffarpur-to-hazaribagh'},
    {'route': 'Purnea to Patna', 'url': 'https://www.redbus.in/bus-tickets/purnea-to-patna'},
    {'route': 'Patna to Araria', 'url': 'https://www.redbus.in/bus-tickets/patna-to-araria'},
    {'route': 'Darbhanga to Patna', 'url': 'https://www.redbus.in/bus-tickets/darbhanga-to-patna'},
    {'route': 'Patna to Saharsa', 'url': 'https://www.redbus.in/bus-tickets/patna-to-saharsa'},
    {'route': 'Motihari to Agra', 'url': 'https://www.redbus.in/bus-tickets/motihari-to-agra'},
    {'route': 'Hajipur to Kathmandu', 'url': 'https://www.redbus.in/bus-tickets/hajipur-to-kathmandu'},
    {'route': 'Kathmandu to Motihari', 'url': 'https://www.redbus.in/bus-tickets/kathmandu-to-motihari'},
    {'route': 'Patna to Forbesganj', 'url': 'https://www.redbus.in/bus-tickets/patna-to-forbesganj'},
    {'route': 'Ranchi to Hajipur', 'url': 'https://www.redbus.in/bus-tickets/ranchi-to-hajipur'},
    {'route': 'Lucknow to Gopalganj', 'url': 'https://www.redbus.in/bus-tickets/lucknow-to-gopalganj'}
]

all_buses = []

for route in urls:
    route_name = route['route']
    route_url = route['url']

    driver.get(route_url)
    # Removed maximize_window() due to the encountered issue
    # driver.maximize_window()
    time.sleep(5)

    

    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        body = driver.find_element(By.TAG_NAME, "body")
        body.send_keys(Keys.END)
        time.sleep(2)  

        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

    try:
        bus_details = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, ".//div[@class='clearfix bus-item']"))
        )
    except Exception as e:
        print(f"Error: {e} for route {route_name}")
        continue

    for bus in bus_details:
        try:
            bus_name = bus.find_element(By.XPATH, ".//div[@class='travels lh-24 f-bold d-color']").text
        except:
            bus_name = "N/A"
        try:
            departure_time = bus.find_element(By.XPATH, ".//div[@class='dp-time f-19 d-color f-bold']").text
        except:
            departure_time = "N/A"
        try:
            reaching_time = bus.find_element(By.XPATH, ".//div[@class='bp-time f-19 d-color disp-Inline']").text
        except:
            reaching_time = "N/A"
        try:
            duration = bus.find_element(By.XPATH, ".//div[@class='dur l-color lh-24']").text
        except:
            duration = "N/A"
        try:
            bus_price = bus.find_element(By.CSS_SELECTOR, "div[class='seat-fare ']").text
        except:
            bus_price = "N/A"
        try:
            seat_availability = bus.find_element(By.CSS_SELECTOR, "div[class='column-eight w-15 fl']").text
        except:
            seat_availability = "N/A"
        try:
            bus_type = bus.find_element(By.XPATH, ".//div[@class='bus-type f-12 m-top-16 l-color evBus']").text
        except:
            bus_type = "N/A"
        try:
            bus_rating = bus.find_element(By.XPATH, ".//div[@class='rating-sec lh-24']").text
        except:
            bus_rating = "N/A"

        all_buses.append({
            "Bus Name": bus_name,
            "Departure Time": departure_time,
            "Reaching Time": reaching_time,
            "Duration": duration,
            "Bus Price": bus_price,
            "Seat Availability": seat_availability,
            "Bus Type": bus_type,
            "Rating": bus_rating,
            "Bus Route": route_name,
            "Bus Route Link": route_url
        })

df = pd.DataFrame(all_buses)
print(df)

driver.quit()
```
