```python
import pandas as pd
import pymysql

# Establish connection
myconnection = pymysql.connect(
    host='127.0.0.1',
    user='root',
    passwd='Sudha16@$',
    database='redbus'
)

# Read CSV file
df = pd.read_csv("C:/Users/SUDHA/Downloads/assam_bus_details.csv")

# Drop rows with missing values
df = df.dropna()

# Generate SQL column definitions from DataFrame dtypes
a = ",".join(
    f"`{i}` {j}".replace("float64", "FLOAT")
               .replace("category", "TEXT")
               .replace("int64", "INT")
               .replace("object", "TEXT")
               .replace("bool", "TEXT")
    for i, j in zip(df.columns, df.dtypes)
)

# Table name
table_name = "assam_bus_details"

# Create table
cursor = myconnection.cursor()
cursor.execute(f"CREATE TABLE IF NOT EXISTS {table_name} ({a})")

# Insert data into the table
for i in range(len(df)):
    placeholders = ", ".join(["%s"] * len(df.columns))
    columns = ", ".join([f"`{col}`" for col in df.columns])
    sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
    cursor.execute(sql, tuple(df.iloc[i]))

myconnection.commit()


```
