```python
#TSRTC Bus Routes 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

driver = webdriver.Chrome()


time.sleep(5)

driver.get('https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile')


wait = WebDriverWait(driver, 10)


body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
body.send_keys(Keys.PAGE_DOWN)


bus_routes = []

for page_number in range(1, 4):
    try:
      
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[class='route']")))

   
        transport_name_elements = driver.find_elements(By.CSS_SELECTOR, "a[class='route']")

      
        for element in transport_name_elements:
            route = element.text.strip()
            if route and route != 'bus_routes':
                bus_routes.append(route)

        if page_number < 3:
       
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            
            
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            
            driver.maximize_window()
            actions = ActionChains(driver)                                       
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
         
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route in bus_routes:
    print(route)

driver.quit()
```

```python
#Assam State Transport Corporation (ASTC) Bus Routes link

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


driver = webdriver.Chrome()

time.sleep(5)


driver.get('https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile')

wait = WebDriverWait(driver, 10)
body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))


driver.maximize_window()

body.send_keys(Keys.PAGE_DOWN)
time.sleep(2)


bus_route_links = []


for page_number in range(1, 4):
    try:
       
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='route_link']")))

        transport_name_elements = driver.find_elements(By.XPATH, "//div[@class='route_link']")
        
      
        for element in transport_name_elements:
            link_element = element.find_element(By.TAG_NAME, 'a')
            route_link = link_element.get_attribute('href')  
            bus_route_links.append(route_link)

       
        if page_number < 3:
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route_link in bus_route_links:
    print(f"Link: {route_link}")


driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.implicitly_wait(10)

urls = [
    {'route':'Khammam to Hyderabad','url':'https://www.redbus.in/bus-tickets/khammam-to-hyderabad'},
    {'route':'Hyderabad to Vijayawada','url':'https://www.redbus.in/bus-tickets/hyderabad-to-vijayawada'},
    {'route':'Hyderabad to Khammam','url':'https://www.redbus.in/bus-tickets/hyderabad-to-khammam'},
    {'route':'Hyderabad to Srisailam','url':'https://www.redbus.in/bus-tickets/hyderabad-to-srisailam'},
    {'route':'Karimnagar to Hyderabad','url':'https://www.redbus.in/bus-tickets/karimnagar-to-hyderabad'},
    {'route':'Hyderabad to Nirmal','url':'https://www.redbus.in/bus-tickets/hyderabad-to-nirmal'},
    {'route':'Hyderabad to Mancherial','url':'https://www.redbus.in/bus-tickets/hyderabad-to-mancherial'},
    {'route':'Hyderabad to Adilabad','url':'https://www.redbus.in/bus-tickets/hyderabad-to-adilabad'},
    {'route':'Hyderabad to Karimnagar','url':'https://www.redbus.in/bus-tickets/hyderabad-to-karimnagar'},
    {'route':'Kothagudem to Hyderabad','url':'https://www.redbus.in/bus-tickets/kothagudem-to-hyderabad'},
    {'route':'Guntur (Andhra Pradesh) to Hyderabad','url':'https://www.redbus.in/bus-tickets/guntur-to-hyderabad'},
    {'route':'Godavarikhani to Hyderabad','url':'https://www.redbus.in/bus-tickets/godavarikhani-to-hyderabad'},
    {'route':'Kodad to Hyderabad','url':'https://www.redbus.in/bus-tickets/kodad-to-hyderabad'},
    {'route':'Hyderabad to Bhadrachalam','url':'https://www.redbus.in/bus-tickets/hyderabad-to-bhadrachalam'},
    {'route':'Hyderabad to Kothagudem','url':'https://www.redbus.in/bus-tickets/hyderabad-to-kothagudem'},
    {'route':'Jagityal to Hyderabad','url':'https://www.redbus.in/bus-tickets/jagityal-to-hyderabad'},
    {'route':'Hyderabad to Kurnool','url':'https://www.redbus.in/bus-tickets/hyderabad-to-kurnool'},
    {'route':'Hyderabad to Sathupally','url':'https://www.redbus.in/bus-tickets/hyderabad-to-sathupalli'},
    {'route':'Hyderabad to Armoor','url':'https://www.redbus.in/bus-tickets/hyderabad-to-armoor'},
    {'route':'Hyderabad to Godavarikhani','url':'https://www.redbus.in/bus-tickets/hyderabad-to-godavarikhani'},
    {'route':'Hyderabad to Anantapur (andhra pradesh)','url':'https://www.redbus.in/bus-tickets/hyderabad-to-ananthapur'},
    {'route':'Hyderabad to Warangal','url':'https://www.redbus.in/bus-tickets/hyderabad-to-warangal'},
    {'route':'Hyderabad to Ongole','url':'https://www.redbus.in/bus-tickets/hyderabad-to-ongole'},
    {'route':'Hyderabad to Guntur (Andhra Pradesh)','url':'https://www.redbus.in/bus-tickets/hyderabad-to-guntur'},
    {'route':'Kadapa to Hyderabad','url':'https://www.redbus.in/bus-tickets/kadapa-to-hyderabad'},
    {'route':'Hyderabad to Tirupati','url':'https://www.redbus.in/bus-tickets/hyderabad-to-tirupathi'}
]

all_buses = []

for route in urls:
    route_name = route['route']
    route_url = route['url']

    driver.get(route_url)
    # Removed maximize_window() due to the encountered issue
    # driver.maximize_window()
    time.sleep(5)

    try:
        view_buses_buttons = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div[class='button']"))
        )
        for button in view_buses_buttons:
            try:
                driver.execute_script("arguments[0].scrollIntoView(true);", button)
                WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, "div[class='button']"))
                )
                driver.execute_script("arguments[0].click();", button)
                time.sleep(2)
            except Exception as e:
                print(f"Error clicking 'View Buses' button: {e}")
                continue
    except Exception as e:
        print(f"Error finding 'View Buses' buttons for route {route_name}: {e}")
        continue    

    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        body = driver.find_element(By.TAG_NAME, "body")
        body.send_keys(Keys.END)
        time.sleep(2)  

        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

    try:
        bus_details = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, ".//div[@class='clearfix bus-item']"))
        )
    except Exception as e:
        print(f"Error: {e} for route {route_name}")
        continue

    for bus in bus_details:
        try:
            bus_name = bus.find_element(By.XPATH, ".//div[@class='travels lh-24 f-bold d-color']").text
        except:
            bus_name = "N/A"
        try:
            departure_time = bus.find_element(By.XPATH, ".//div[@class='dp-time f-19 d-color f-bold']").text
        except:
            departure_time = "N/A"
        try:
            reaching_time = bus.find_element(By.XPATH, ".//div[@class='bp-time f-19 d-color disp-Inline']").text
        except:
            reaching_time = "N/A"
        try:
            duration = bus.find_element(By.XPATH, ".//div[@class='dur l-color lh-24']").text
        except:
            duration = "N/A"
        try:
            bus_price = bus.find_element(By.CSS_SELECTOR, "div[class='seat-fare ']").text
        except:
            bus_price = "N/A"
        try:
            seat_availability = bus.find_element(By.CSS_SELECTOR, "div[class='column-eight w-15 fl']").text
        except:
            seat_availability = "N/A"
        try:
            bus_type = bus.find_element(By.XPATH, ".//div[@class='bus-type f-12 m-top-16 l-color evBus']").text
        except:
            bus_type = "N/A"
        try:
            bus_rating = bus.find_element(By.XPATH, ".//div[@class='rating-sec lh-24']").text
        except:
            bus_rating = "N/A"

        all_buses.append({
            "Bus Name": bus_name,
            "Departure Time": departure_time,
            "Reaching Time": reaching_time,
            "Duration": duration,
            "Bus Price": bus_price,
            "Seat Availability": seat_availability,
            "Bus Type": bus_type,
            "Rating": bus_rating,
            "Bus Route": route_name,
            "Bus Route Link": route_url
        })

df = pd.DataFrame(all_buses)
print(df)

driver.quit()
```
