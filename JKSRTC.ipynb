```python
#jksrtc
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

driver = webdriver.Chrome()


time.sleep(5)

driver.get('https://www.redbus.in/online-booking/jksrtc')


wait = WebDriverWait(driver, 10)


body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
body.send_keys(Keys.PAGE_DOWN)


bus_routes = []

for page_number in range(1, 3):
    try:
      
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a[class='route']")))

   
        transport_name_elements = driver.find_elements(By.CSS_SELECTOR, "a[class='route']")

      
        for element in transport_name_elements:
            route = element.text.strip()
            if route and route != 'bus_routes':
                bus_routes.append(route)

        if page_number < 2:
       
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            
            
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            
            driver.maximize_window()
            actions = ActionChains(driver)                                       
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
         
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route in bus_routes:
    print(route)

driver.quit()
```

```python
#JKSRTC BUS ROUTE LINK
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


driver = webdriver.Chrome()

time.sleep(5)


driver.get('https://www.redbus.in/online-booking/jksrtc')

wait = WebDriverWait(driver, 10)
body = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))


driver.maximize_window()

body.send_keys(Keys.PAGE_DOWN)
time.sleep(2)


bus_route_links = []


for page_number in range(1, 3):
    try:
       
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='route_link']")))

        transport_name_elements = driver.find_elements(By.XPATH, "//div[@class='route_link']")
        
      
        for element in transport_name_elements:
            link_element = element.find_element(By.TAG_NAME, 'a')
            route_link = link_element.get_attribute('href')  
            bus_route_links.append(route_link)

       
        if page_number < 2:
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_number + 1}"]')
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
            next_page_button.click()
            
    except Exception as e:
        print(f"An error occurred: {e}")


for route_link in bus_route_links:
    print(f"Link: {route_link}")


driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.implicitly_wait(10)


urls = [
    {"route": "Jammu (j and k) to Srinagar", "url": "https://www.redbus.in/bus-tickets/jammu-to-srinagar"},
    {"route": "Delhi to Srinagar", "url": "https://www.redbus.in/bus-tickets/delhi-to-srinagar"},
    {"route": "Srinagar to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/srinagar-to-jammu"},
    {"route": "Jammu (j and k) to Haridwar", "url": "https://www.redbus.in/bus-tickets/jammu-to-haridwar"},
    {"route": "Jammu (j and k) to Delhi", "url": "https://www.redbus.in/bus-tickets/jammu-to-delhi"},
    {"route": "Delhi to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/delhi-to-jammu"},
    {"route": "Katra (jammu and kashmir) to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/katra-to-jammu"},
    {"route": "Jammu (j and k) to Katra (jammu and kashmir)", "url": "https://www.redbus.in/bus-tickets/jammu-to-katra"},
    {"route": "Mendhar (J & K) to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/mendhar-j-k-to-jammu"},
    {"route": "Jammu (j and k) to Poonch", "url": "https://www.redbus.in/bus-tickets/jammu-to-poonch"},
    {"route": "Haridwar to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/haridwar-to-jammu"},
    {"route": "Chandigarh to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/chandigarh-to-jammu"},
    {"route": "Jammu (j and k) to Kishtwar", "url": "https://www.redbus.in/bus-tickets/jammu-to-kishtwar"},
    {"route": "Jammu (j and k) to Mendhar (J & K)", "url": "https://www.redbus.in/bus-tickets/jammu-to-mendhar-j-k"},
    {"route": "Jammu (j and k) to Amritsar", "url": "https://www.redbus.in/bus-tickets/jammu-to-amritsar"},
    {"route": "Ajmer to Jammu (j and k)", "url": "https://www.redbus.in/bus-tickets/ajmer-to-jammu"}
]


all_buses = []

for route in urls:
    route_name = route['route']
    route_url = route['url']

    # Open the target URL
    driver.get(route_url)
    driver.maximize_window()
    time.sleep(5)

    # Scroll through the page until the end
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source

        for _ in range(10):  # Adjust the number as needed
            body = driver.find_element(By.TAG_NAME, "body")
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1)  # Wait for a second to allow new content to load

        time.sleep(2)

        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

   
    try:
        bus_details = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, ".//div[@class='clearfix bus-item']"))
        )
    except Exception as e:
        print(f"Error: {e} for route {route_name}")
        continue

    for bus in bus_details:
        try:
            bus_name = bus.find_element(By.CSS_SELECTOR, "div.travels.lh-24.f-bold.d-color").text
        except:
            bus_name = "N/A"
        
        try:
            departure_time = bus.find_element(By.CSS_SELECTOR, "div.dp-time.f-19.d-color.f-bold").text
        except:
            departure_time = "N/A"
        
        try:
            reaching_time = bus.find_element(By.CSS_SELECTOR, "div.bp-time.f-19.d-color.disp-Inline").text
        except:
            reaching_time = "N/A"
        
        try:
            duration = bus.find_element(By.CSS_SELECTOR, "div.dur.l-color.lh-24").text
        except:
            duration = "N/A"
        
        try:
            bus_price = bus.find_element(By.CSS_SELECTOR, "div.seat-fare").text
        except:
            bus_price = "N/A"
        
        try:
            seat_availability = bus.find_element(By.CSS_SELECTOR, "div.column-eight.w-15.fl").text
        except:
            seat_availability = "N/A"
        
        try:
            bus_type = bus.find_element(By.CSS_SELECTOR, "div.bus-type.f-12.m-top-16.l-color.evBus").text
        except:
            bus_type = "N/A"
        
        try:
            bus_rating = bus.find_element(By.CSS_SELECTOR, "div.rating-sec.lh-24").text
        except:
            bus_rating = "N/A"


        all_buses.append({
            "Bus Name": bus_name,
            "Departure Time": departure_time,
            "Reaching Time": reaching_time,
            "Duration": duration,
            "Bus Price": bus_price,
            "Seat Availability": seat_availability,
            "Bus Type": bus_type,
            "Rating": bus_rating,
            "Bus Route": route_name,
            "Bus Route Link": route_url
        })

df = pd.DataFrame(all_buses)
print(df)

driver.quit()
```
